## Create the Firestore collections and subcollections based on your design and populate them with the shopify data.

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
```

#### Get apps by category (collection-category, subcollection-apps)

```python
batch = db.batch()

df = pd.read_csv('/home/jupyter/shopify_firestore/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/shopify_firestore/apps_categories.csv', sep=',', header=0, lineterminator='\n')
apps_cat_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/shopify_firestore/categories.csv', sep=',', header=0, lineterminator='\n')
cat_rows = df.values.tolist()

counter = 0

for cat_row in cat_rows:
    
    category_record = {}
    category_record['id'] = cat_row[0]
    category_record['title'] = cat_row[1]
    
    cat_ref = db.collection('category').document(cat_row[0])
    
    counter += 1
    if counter > 399:
        counter = 0
        batch.commit()
        batch = db.batch()
        
    batch.set(cat_ref, category_record)
    
    for apps_cat_row in apps_cat_rows:
        if cat_row[0] == apps_cat_row[1]:
            
            for apps_row in apps_rows:
                if apps_row[0] == apps_cat_row[0]:
                    apps_record = {}
                    apps_record['id'] = apps_row[0]
                    apps_record['url'] = apps_row[1]
                    apps_record['title'] = apps_row[2]
                    apps_record['developer'] = apps_row[3]
                    apps_record['developer_link'] = apps_row[4]
                    apps_record['icon'] = apps_row[5]
                    apps_record['rating'] = apps_row[6]
                    apps_record['review_count'] = apps_row[7]

            
            apps_ref = cat_ref.collection('apps_s').document(apps_cat_row[0])
            
            counter += 1
            if counter > 399:
                counter = 0
                batch.commit()
                batch = db.batch()
                
            batch.set(apps_ref, apps_record)
    
    batch.commit()
```

#### Get pricing plan details by app (collection-apps, subcollection-pricing plan)

```python
batch = db.batch()

df = pd.read_csv('/home/jupyter/shopify_firestore/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/shopify_firestore/pricing_plans.csv', sep=',', header=0, lineterminator='\n')
pricingp_rows = df.values.tolist()

counter = 0

for apps_row in apps_rows:
    
    apps_record = {}
    apps_record['id'] = apps_row[0]
    apps_record['url'] = apps_row[1]
    apps_record['title'] = apps_row[2]
    apps_record['developer'] = apps_row[3]
    apps_record['developer_link'] = apps_row[4]
    apps_record['icon'] = apps_row[5]
    apps_record['rating'] = apps_row[6]
    apps_record['review_count'] = apps_row[7]
    
    apps_ref = db.collection('apps').document(apps_row[0])
    
    counter += 1
    if counter > 399:
        counter = 0
        batch.commit()
        batch = db.batch()
    
    batch.set(apps_ref, apps_record)
    
    for pricingp_row in pricingp_rows:
        
        if apps_row[0] == pricingp_row[1]:        
            pricingp_record = {}
            pricingp_record['id'] = pricingp_row[0]
            pricingp_record['app_id'] = pricingp_row[1]
            pricingp_record['title'] = pricingp_row[2]
            pricingp_record['price'] = pricingp_row[3]
            
            pricingp_ref = apps_ref.collection('pricing_plans').document(pricingp_row[0])
            
            counter += 1
            if counter > 399:
                counter = 0
                batch.commit()
                batch = db.batch()
        
            batch.set(pricingp_ref, pricingp_record)
    
    batch.commit()
```

## List the 10 apps with the highest number of reviews (based on apps.review_count). Return the id, title, developer, rating and reviews_count of those apps. Order the results by reviews_count in descending order.

```python
apps = db.collection('apps').order_by('review_count', direction=firestore.Query.DESCENDING).limit(10)
stream = apps.stream()
app_dict = {}

count = 0

for row in stream:
    newrow = row.to_dict()
    app_dict.update(newrow)
    count += 1
    print("top " + str(count))
    print(app_dict['id'], app_dict['title'], app_dict['developer'], app_dict['rating'], app_dict['review_count'])
```
