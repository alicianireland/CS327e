## Write 3 queries that use an INNER JOIN between two or more tables. The queries should also include a WHERE clause and ORDER BY clause. Feel free to include a LIMIT clause if the output returns > 10 rows. Add a short Markdown comment above each SQL statement.

```python
!psql -c "select s.first_name, s.last_name, s.email, st.manager_staff_id from pagila.staff s join pagila.store st on s.store_id = st.store_id where username IS NOT NULL order by first_name, last_name"
```

This inner join statement chose first name, last name, and email columns from staff table and store manager's staff id from store table, joined staff table and store table using store id column that the tables share, and listed the result of this inner join where the username is not null in order by first name then last name (if there are people with the same first names). Alias on table names were used.

```python
!psql -c "select * from pagila.inventory join pagila.store on pagila.store.store_id = pagila.inventory.store_id where pagila.store.store_id = 1 order by inventory_id limit 10"
```

This inner join statement joined all columns of store and inventory tables based on store id column that the tables share and showed the results that has store id of 1 in order by inventory id (ascending). The joined result was limited to 10 because the output consisted of more than 10 rows. 

```python
!psql -c "select * from pagila.film_actor fa join pagila.film_category fc on fa.film_id = fc.film_id where actor_id in (2,3) order by category_id limit 10"
```

This inner join statement joined all columns of film_actor table and film_category table based on the film_id column that is shared between both tables and produced output where the actor id is either 2 or 3 and ordered them by category_id. Because the produced output was more than 10 rows, the output was limited to 10 rows. Alias on table names were used.

##  Write 3 queries that use an OUTER JOIN between two or more tables. The queries should also include a WHERE clause and ORDER BY clause. Feel free to include a LIMIT clause if the output returns > 10 rows. Add a short Markdown comment above each SQL statement.

```python
!psql -c "select inv.inventory_id, inv.film_id, st.store_id, st.last_update from pagila.inventory inv right join pagila.store st on inv.store_id = st.store_id where film_id IS NOT NULL order by inventory_id limit 10"
```

This is a right outer join statement of inventory and store tables' inventory id, film id, store id, and last update column based on the store id column which is shared between both tables. The resulting outer join output shows rows where film_id is not a null value and the result is ordered by inventory id (ascending). The results returns all data from store table but only those from inventory table that had match with store table data and the results were limited to 10 rows.

```python
!psql -c "select * from pagila.city left join pagila.country on pagila.city.country_id = pagila.country.country_id where city_id in (3,5,8,17,20,150) order by city"
```

This is a left outer join statement joins city and country tables in pagila dataset based on country id column that both tables share. The resulting output of this left outer join statement is shown where city_id is among 3, 5, 8, 17, 20, and 150 and is ordered by the name of the city(ascneding). Because this is a left join, all records on city table will appear on the output whereas from country table, only the results that had a matching country id with city table will show up.

```python
!psql -c "select adr.address_id, adr.address, adr.district, city.city_id, adr.postal_code, city.city, city.country_id from pagila.address adr full join pagila.city city on adr.city_id = city.city_id where postal_code IS NOT NULL order by country_id, city limit 10"
```

This is a full outer join statement of address and city table with address id, address, district, city id, postal code, city, and country id columns based on city id column they share. This returned all unmatched rows from both city and address tables and the output is shown where postal code is not a null value and is listed in order by country id and city if the country id overlaps. 

## Write 1 query that uses a self join on any table. Feel free to include a LIMIT clause if the output returns > 10 rows. Add a short Markdown comment above each SQL statement.

```python
!psql -c "select a1.first_name, a1.last_name, a2.last_name, a2.actor_id from pagila.actor a1 join pagila.actor a2 on a1.actor_id != a2.actor_id and a1.last_name = a2.last_name and a1.first_name = a2.first_name where a1.actor_id < 120 order by a2.actor_id"
```

This self join statement joins actor table with itself based on seeing if different actors under different actor id has same last names and first names and produced output based on if the actor's actor id is less than 120 or not, ordered by the actor id. So essentially, we are trying to see if any actors have the same first name and last name (assuming actor id cannot be assigned twice to the same person).
